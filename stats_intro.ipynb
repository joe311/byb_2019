```python
import numpy as np
import scipy

%matplotlib notebook

from matplotlib import pyplot as plt, gridspec
import seaborn.apionly as sns
```

```python
npoints = 100
scale = .2
means = np.random.normal(loc=0, scale=scale, size=(npoints,100000)).mean(axis=0)
means = means*np.sqrt(npoints)/scale
means.shape
```

```python
plt.figure()
plt.hist(means, 100);
```

```python
#The sampled and analytically calculated distributions are close
plt.figure()
sns.kdeplot(means, label='sampled mean distribution', lw=3)

x = np.linspace(-5, 5, 1000)
plt.plot(x, scipy.stats.t.pdf(x, df=npoints), 'k--', label='true t-dist', alpha=.5)
plt.legend(loc='upper right')
```
